```python
nqcpi = qcpi.iloc[:,1].to_numpy()
plt.figure(figsize=(12,7))
plt.plot(nqcpi)
plt.show()
```

```python
qinfln = (nqcpi[4:] - nqcpi[:-4])/nqcpi[:-4]
plt.figure(figsize=(12,7))
plt.plot(qinfln)
plt.show()
```

```python
qinfln.mean(), qinfln.std()
```

```python
plt.figure(figsize=(12,7))
plt.hist(qinfln,bins=30)
plt.show()
```

```python
data = np.abs(qinfln[qinfln>0])
results = pl.Fit(data)
a = results.alpha
m = results.xmin
a, m
```

```python
### complementary cdf plot Ac ###
plt.figure(figsize=(12,8))
# x values 
x = np.sort(data)
# complementary cdf (ccdf)
cdf = np.arange(1,len(data)+1)/(len(data))
ccdf = 1 - cdf
plt.scatter(x, ccdf, color='k', s=30, label='CCDF')
# power law fit: theoretic
# => rescale to start fit from cut off
index = np.where(x == m)[0][0]
rescale = ccdf[index]
power_law_fit = np.where(x >= m, np.power((m)/x,a-1)*rescale, np.nan)
plt.plot(x, power_law_fit, color='red', linewidth=2, label='Power-Law Fit')
# power law cut off (mF)
plt.axvline(m, color='k', linestyle='--', label=r'$m$'f' = {round(m)}')
# log-log axis
plt.loglog()
# legend
plt.legend(fontsize=20, loc='lower left')
# tick size
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
# plt.savefig(f"{filepath}/Ac_ccdf_{year}.pdf")
plt.show()
```

```python
results.distribution_compare('power_law', 'lognormal')
```

```python
nrgdp = rgdp['RealGDP'].to_numpy()
qrgdp_growth = (nrgdp[4:] - nrgdp[:-4])/nrgdp[:-4]
plt.figure(figsize=(12,7))
plt.plot(qrgdp_growth)
plt.show()
```

```python
avg_growth_rate = (nrgdp[-1]/nrgdp[0])**(1/(nrgdp.shape[0]/4)) - 1
avg_growth_rate, qrgdp_growth.std()
```

```python
# estimated distribution function
sorted_qrgdp_growth = np.sort(qrgdp_growth)
edf = np.arange(1, len(sorted_qrgdp_growth)+1)/len(sorted_qrgdp_growth)

# normal fit 
mean = np.mean(sorted_qrgdp_growth)
std = np.std(sorted_qrgdp_growth)
norm_cdf = st.norm.cdf(sorted_qrgdp_growth, mean, std)

# cauchy fit
median = np.median(sorted_qrgdp_growth)
scale = (np.quantile(sorted_qrgdp_growth,0.75) - np.quantile(sorted_qrgdp_growth, 0.25))/2
cauchy_cdf = st.cauchy.cdf(sorted_qrgdp_growth, median, scale)

# laplace fit 
median = np.median(sorted_qrgdp_growth)
sale = (np.quantile(sorted_qrgdp_growth,0.75) - np.quantile(sorted_qrgdp_growth, 0.25))/2
laplace_cdf = st.laplace.cdf(sorted_qrgdp_growth, median, sale)

# generalized normal fit 
median = np.median(sorted_qrgdp_growth)
sale = (np.quantile(sorted_qrgdp_growth,0.75) - np.quantile(sorted_qrgdp_growth, 0.25))/2
shape = 0.8
gennorm_cdf = st.gennorm.cdf(sorted_qrgdp_growth, shape, median, sale)


# plot results 
plt.figure(figsize=(12,7))
plt.scatter(x=sorted_qrgdp_growth, y=edf,color='steelblue',alpha=0.7)
plt.plot(sorted_qrgdp_growth, norm_cdf, color='g')
# plt.plot(sorted_qrgdp_growth, cauchy_cdf, color='b')
plt.plot(sorted_qrgdp_growth, laplace_cdf, color='y')
plt.plot(sorted_qrgdp_growth, gennorm_cdf, color='r')
plt.show()
```
